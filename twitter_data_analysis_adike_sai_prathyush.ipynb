```python
#removing some more patterns which remained after texthero was used by checking individual tweets
df['chatter']= df['chatter'].str.replace("ðyœðyœ", "")
df['chatter']= df['chatter'].str.replace('[^a-zA-Z]', ' ')
df['chatter']= df['chatter'].str.replace(r'\b(\w{1,3})\b', '')
df['chatter']= df['chatter'].str.strip()
df['tweets']= df['tweets'].str.replace("ðyœðyœ", "")
df['tweets']= df['tweets'].str.replace('[^a-zA-Z]', ' ')
df['tweets']= df['tweets'].str.replace(r'\b(\w{1,3})\b', '')
df['tweets']= df['tweets'].str.strip()
```

```python
# using tweet hero to remove un wanted gaps emerged by removing patterns
df['chatter'] = hero.clean(df['chatter'])
df['tweets'] = hero.clean(df['tweets'])
```

```python
from nltk.corpus import stopwords
stop = stopwords.words('english')
df['tweets'] = df['tweets'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))
df['chatter'] = df['chatter'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))
```

```python
#using textblob to convert text into sentiment
from textblob import TextBlob
df['chatter_sentiment_subjectivity'] = df['chatter'].apply(lambda tweet: TextBlob(tweet).sentiment.subjectivity)
df['chatter_sentiment_polarity'] = df['chatter'].apply(lambda tweet: TextBlob(tweet).sentiment.polarity)
df['tweets_sentiment_subjectivity'] = df['tweets'].apply(lambda tweet: TextBlob(tweet).sentiment.subjectivity)
df['tweets_sentiment_polarity'] = df['tweets'].apply(lambda tweet: TextBlob(tweet).sentiment.polarity)
```

```python
#dropping chatter and tweets columns as there is no more use of them
df=df.drop(['chatter'], axis = 1)
df=df.drop(['tweets'], axis = 1)
```

```python
X=train.drop(columns=['rank on the following day (dpv1)',
 'rank 2 days later            ( dpv2)',
 'rank 3 days later          (dpv3)','avg_tweet_retweets','count_tweets','avg_tweet_likes'])
```

```python
X_test=test.drop(columns=['rank on the following day (dpv1)',
 'rank 2 days later            ( dpv2)',
 'rank 3 days later          (dpv3)','avg_tweet_retweets','count_tweets','avg_tweet_likes'])
```

```python
Y=train[['rank on the following day (dpv1)'
       ,'rank 2 days later            ( dpv2)','rank 3 days later          (dpv3)']]
```

```python
Y_test=test[['rank on the following day (dpv1)'
       ,'rank 2 days later            ( dpv2)','rank 3 days later          (dpv3)']]
```

```python
from sklearn.ensemble import BaggingRegressor
from sklearn.pipeline import make_pipeline
from sklearn.multioutput import MultiOutputRegressor
import xgboost as xgb
from sklearn.preprocessing import MinMaxScaler
reg = make_pipeline(MinMaxScaler(),BaggingRegressor(base_estimator=xgb.XGBRegressor(objective ='reg:linear', colsample_bytree = 0.3, learning_rate = 0.5,max_depth = 30, alpha = 10, n_estimators = 100) ,n_estimators=100, random_state=2))
multilabel_model = MultiOutputRegressor(reg)
multilabel_model.fit(X,Y)
```

```python
Y_pred = multilabel_model.predict(X_test)
from numpy import sqrt 
import sklearn.metrics as metrics
print('rmse',sqrt(metrics.mean_squared_error(Y_test,Y_pred)))
```
